# Analyzing air quality in Gdansk during 30-days period

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("./openaq_wrzeszcz.csv", parse_dates=True, 
                                   usecols = ["local", "parameter", "value", "unit"])
air_quality["parameter"].value_counts()
```

```python
air_quality.set_index('local', inplace = True)
air_quality = air_quality.pivot(columns="parameter", values="value")
```

```python
air_quality.plot(figsize=(30,10))
```

```python
air_quality.drop(columns=['co']).plot(figsize=(30,10))
```

```python
air_quality['no2'].plot(figsize=(30,10))
```

```python
air_quality['o3'].plot(figsize=(30,10))
```

# Analyzing air quality in Gdansk during 30-days period

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("./openaq_wrzeszcz.csv", parse_dates=True, 
                                   usecols = ["local", "parameter", "value", "unit"])
air_quality["parameter"].value_counts()
```

```python
grouped = air_quality.groupby('parameter')
```

```python
grouped.plot(xlabel="date")
```
